## 1. Conversión EDF a CSV

```python
# Ruta donde están los archivos .edf
ruta_entrada = r"C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\dataset_arithmetic"
ruta_salida = os.path.join(ruta_entrada, "Resultados")

# Crear carpeta de salida si no existe
os.makedirs(ruta_salida, exist_ok=True)

# Procesar todos los archivos .edf en la carpeta
for archivo in os.listdir(ruta_entrada):
    if archivo.lower().endswith('.edf'):
        ruta_archivo = os.path.join(ruta_entrada, archivo)
        print(f"Procesando: {archivo}")

        try:
            f = pyedflib.EdfReader(ruta_archivo)
            n_canales = f.signals_in_file
            etiquetas = f.getSignalLabels()
            nsamples = f.getNSamples()
            frecuencias = f.getSampleFrequencies()

            if len(set(nsamples)) == 1:
                print(f"Todos los canales tienen {nsamples[0]} muestras.")
            else:
                print(f"Advertencia: Los canales tienen diferente número de muestras en {archivo}")

            datos = {}
            for i in range(n_canales):
                datos[etiquetas[i]] = f.readSignal(i)

            f.close()

            df = pd.DataFrame(datos)

            # Insertar columna de tiempo (asume misma frecuencia en todos los canales)
            tiempo = np.arange(df.shape[0]) / frecuencias[0]
            df.insert(0, 'Tiempo (s)', tiempo)

            # Guardar como CSV con nombre del archivo original
            nombre_csv = os.path.splitext(archivo)[0] + ".csv"
            ruta_csv = os.path.join(ruta_salida, nombre_csv)
            df.to_csv(ruta_csv, index=False, decimal='.', sep=';')

            print(f"Guardado: {ruta_csv}\n")

        except Exception as e:
            print(f"Error procesando {archivo}: {e}\n")
```

```python
# Rutas
relaxed_source = r'C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Relaxed_Complete'
stress_source = r'C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Stress_Complete'
relaxed_target = r'C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Relaxed'
stress_target = r'C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Stress'

# Crear carpetas de salida
os.makedirs(relaxed_target, exist_ok=True)
os.makedirs(stress_target, exist_ok=True)

# Lista de archivos a excluir
excluir = {
    'Subject04_1.csv', 'Subject04_2.csv',
    'Subject31_1.csv', 'Subject31_2.csv'
}

# ---------- PROCESAMIENTO DE RELAXED ----------
print("\nProcesando archivos Relaxed...")
for file in os.listdir(relaxed_source):
    if not file.endswith('.csv') or file in excluir:
        continue

    path = os.path.join(relaxed_source, file)
    df = pd.read_csv(path, sep=';')

    # Filtrar por Tiempo <= 180
    df_filtered = df[df['Tiempo (s)'] <= 180]
    
    # Guardar
    df_filtered.to_csv(os.path.join(relaxed_target, file), sep=';', index=False)
    print(f"[RELAXED] {file}: {len(df_filtered)} filas guardadas.")

# ---------- PROCESAMIENTO DE STRESS ----------
print("\nProcesando archivos Stress...")
for file in os.listdir(stress_source):
    if not file.endswith('.csv') or file in excluir:
        continue

    path = os.path.join(stress_source, file)
    df = pd.read_csv(path, sep=';')

    df_base = df[df['Tiempo (s)'] <= 60].copy()

    if df_base.empty or len(df_base) < 2:
        print(f"[STRESS] {file} no tiene suficientes datos hasta 60s. Se omite.")
        continue

    # Calcular incremento temporal (step)
    tiempos = df_base['Tiempo (s)'].values
    tiempo_step = round(tiempos[1] - tiempos[0], 6)
    tiempo_actual = tiempos[-1]

    # Columnas EEG (excluyendo Tiempo)
    eeg_cols = [col for col in df.columns if col != 'Tiempo (s)']
    base_values = df_base[eeg_cols].values

    datos_generados = []
    base_index = 0

    while tiempo_actual < 180:
        fila_original = base_values[base_index % len(base_values)]
        escala = np.random.uniform(0.97, 1.03)
        fila_escalada = fila_original * escala

        tiempo_actual = round(tiempo_actual + tiempo_step, 6)
        nueva_fila = dict(zip(eeg_cols, fila_escalada))
        nueva_fila['Tiempo (s)'] = tiempo_actual

        datos_generados.append(nueva_fila)
        base_index += 1

    df_sintetico = pd.DataFrame(datos_generados)
    df_final = pd.concat([df_base, df_sintetico], ignore_index=True)
    df_final = df_final[df.columns]  # asegura el orden de columnas
    df_final.to_csv(os.path.join(stress_target, file), sep=';', index=False)

    print(f"[STRESS] {file}: {len(df_base)} reales + {len(df_sintetico)} generadas = {len(df_final)} filas.")
```

## 5. Visualización de coeficientes wavelet y FFT

```python
def plot_wavelet_segment(coefs_r, coefs_s, niveles, canal, subject, fs, idx_start, idx_end, parte, output_dir):
    num_niveles = idx_end - idx_start
    plt.figure(figsize=(14, 3 * num_niveles))

    for i in range(num_niveles):
        idx = idx_start + i
        coef_r = coefs_r[idx]
        coef_s = coefs_s[idx]

        nivel = 8 if idx == 0 else 9 - idx
        fs_actual = fs / (2 ** nivel)

        plt.subplot(num_niveles, 2, 2 * i + 1)
        plt.plot(coef_r, label='Relaxed', alpha=0.7)
        plt.plot(coef_s, label='Stress', alpha=0.7)
        plt.title(f'{niveles[idx]} - Coeficientes')
        plt.grid(True)
        plt.legend()

        fft_r = np.abs(np.fft.fft(coef_r))[:len(coef_r) // 2]
        fft_s = np.abs(np.fft.fft(coef_s))[:len(coef_s) // 2]
        freqs = np.fft.fftfreq(len(coef_r), d=1 / fs_actual)[:len(coef_r) // 2]

        plt.subplot(num_niveles, 2, 2 * i + 2)
        plt.plot(freqs, fft_r, label='Relaxed', alpha=0.7)
        plt.plot(freqs, fft_s, label='Stress', alpha=0.7)
        plt.title(f'{niveles[idx]} - Espectro de Frecuencia (fs={fs_actual:.2f} Hz)')
        plt.xlabel("Frecuencia (Hz)")
        plt.ylabel("Magnitud")
        plt.grid(True)
        plt.legend()

    plt.tight_layout()
    plt.suptitle(f'Comparación Wavelet - Subject {str(subject).zfill(2)} - Canal: {canal} (Parte {parte})',
                 fontsize=16, y=1.02)

    filename = f'Sujeto_{str(subject).zfill(2)}_Canal_{canal.replace(" ", "_")}_Parte_{parte}.png'
    os.makedirs(output_dir, exist_ok=True)
    path = os.path.join(output_dir, filename)
    plt.savefig(path, bbox_inches='tight', dpi=300)
    print(f">> Imagen guardada en: {path}")
    plt.close()

def wavelet_comparativa_plot_split(signal_relaxed, signal_stress, subject, canal, fs=500):
    coef_relaxed = pywt.wavedec(signal_relaxed, 'db4', level=8)
    coef_stress = pywt.wavedec(signal_stress, 'db4', level=8)
    niveles = ['cA8', 'cD8', 'cD7', 'cD6', 'cD5', 'cD4', 'cD3', 'cD2', 'cD1']
    output_dir = r'C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Imagenes_EEG\Sujeto 10'

    # Parte 1: índices 0 al 3 (4 niveles)
    plot_wavelet_segment(coef_relaxed, coef_stress, niveles, canal, subject, fs, idx_start=0, idx_end=4, parte=1, output_dir=output_dir)

    # Parte 2: índices 4 al 8 (5 niveles)
    plot_wavelet_segment(coef_relaxed, coef_stress, niveles, canal, subject, fs, idx_start=4, idx_end=9, parte=2, output_dir=output_dir)

wavelet_comparativa_plot_split(relaxed_df[canal].values, stress_df[canal].values, subject, canal)
```

## 6. Extracción de entropía en las tres estapas (DWT, DWT+FFT, DWT+FFT+PSD)

```python
# Ruta de salida
OUTPUT_PATH = r"C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados"

# Canales disponibles
canales = ['EEG Fp1', 'EEG Fp2', 'EEG F3', 'EEG F4', 'EEG F7', 'EEG F8', 'EEG T3', 'EEG T4',
           'EEG C3', 'EEG C4', 'EEG T5', 'EEG T6', 'EEG P3', 'EEG P4', 'EEG O1', 'EEG O2',
           'EEG Fz', 'EEG Cz', 'EEG Pz', 'EEG A2-A1']

# Rutas
RELAXED_PATH = r"C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Relaxed"
STRESS_PATH = r"C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Stress"

# --- Leer archivo por sujeto y condición ---
def leer_archivo(subject_num, estado):
    carpeta = RELAXED_PATH if estado == 'Relaxed' else STRESS_PATH
    sufijo = '_1.csv' if estado == 'Relaxed' else '_2.csv'
    nombre_archivo = f"Subject{str(subject_num).zfill(2)}{sufijo}"
    ruta = os.path.join(carpeta, nombre_archivo)
    return pd.read_csv(ruta, sep=';')

# --- Entropía de Shannon (corregida: distribución discreta) ---
def shannon_entropy(signal, bins=256):
    hist, _ = np.histogram(signal, bins=bins)
    prob = hist / np.sum(hist)
    prob = prob[prob > 0]
    return -np.sum(prob * np.log2(prob))

# --- Análisis completo para todos los sujetos ---
def wavelet_entropia_todos_los_sujetos(fs=500):
    resultados = []
    niveles = ['cA8', 'cD8', 'cD7', 'cD6', 'cD5', 'cD4', 'cD3', 'cD2', 'cD1']
    sujetos_excluir = {4, 31}

    for subject in range(36):
        if subject in sujetos_excluir:
            print(f"[SKIP] Sujeto {subject} excluido.")
            continue

        try:
            relaxed_df = leer_archivo(subject, 'Relaxed')
            stress_df = leer_archivo(subject, 'Stress')
            print(f"✅ Procesando sujeto {subject}")
        except FileNotFoundError as e:
            print(f"[ERROR] {e}")
            continue

        for canal in canales:
            if canal not in relaxed_df.columns or canal not in stress_df.columns:
                print(f"[!] Canal {canal} no disponible en sujeto {subject}. Se omite.")
                continue

            signal_relaxed = relaxed_df[canal].values
            signal_stress = stress_df[canal].values

            coef_relaxed = pywt.wavedec(signal_relaxed, 'db4', level=8)
            coef_stress = pywt.wavedec(signal_stress, 'db4', level=8)

            for i in range(len(coef_relaxed)):
                coef_r = coef_relaxed[i]
                coef_s = coef_stress[i]

                nivel = 8 if i == 0 else 9 - i
                fs_actual = fs / (2 ** nivel)

                # FFT
                fft_r = np.abs(np.fft.fft(coef_r))[:len(coef_r) // 2]
                fft_s = np.abs(np.fft.fft(coef_s))[:len(coef_s) // 2]

                # PSD con Welch aplicado a la FFT
                _, psd_r = welch(fft_r, fs=fs_actual, nperseg=min(256, len(fft_r)))
                _, psd_s = welch(fft_s, fs=fs_actual, nperseg=min(256, len(fft_s)))

                # Cálculo de entropías
                ent_coef_r = shannon_entropy(coef_r)
                ent_fft_r = shannon_entropy(fft_r)
                ent_psd_r = shannon_entropy(psd_r)

                ent_coef_s = shannon_entropy(coef_s)
                ent_fft_s = shannon_entropy(fft_s)
                ent_psd_s = shannon_entropy(psd_s)

                resultados.append([subject, canal, niveles[i], 'Relaxed', ent_coef_r, ent_fft_r, ent_psd_r])
                resultados.append([subject, canal, niveles[i], 'Stress', ent_coef_s, ent_fft_s, ent_psd_s])

        print(f"📦 Sujeto {subject} procesado.\n")

    # Guardado
    if resultados:
        df_resultados = pd.DataFrame(resultados, columns=[
            'Sujeto', 'Canal', 'Coeficiente', 'Condicion',
            'Entropia Coeficiente', 'Entropia FFT', 'Entropia PSD'
        ])

        full_path = os.path.join(OUTPUT_PATH, "wavelet_entropy_todos_los_sujetos.csv")
        df_resultados.to_csv(full_path, sep=';', index=False)
        print(f"\n✅ CSV combinado guardado: {full_path}")
        print(f"📊 Total de registros: {len(df_resultados)}")
    else:
        print("⚠️ No se generaron resultados. Revisa los archivos de entrada y nombres de canales.")

# Ejecutar
wavelet_entropia_todos_los_sujetos()
```

## 7. Visualización de coeficientes, FFT y PSD

```python
def plot_fft_psd_segment(coefs_r, coefs_s, niveles, canal, subject, fs, idx_start, idx_end, parte, output_dir):
    num_niveles = idx_end - idx_start
    plt.figure(figsize=(14, 3 * num_niveles))

    for i in range(num_niveles):
        idx = idx_start + i
        coef_r = coefs_r[idx]
        coef_s = coefs_s[idx]

        nivel = 8 if idx == 0 else 9 - idx
        fs_actual = fs / (2 ** nivel)

        # FFT
        fft_r = np.abs(np.fft.fft(coef_r))[:len(coef_r) // 2]
        fft_s = np.abs(np.fft.fft(coef_s))[:len(coef_s) // 2]
        freqs = np.fft.fftfreq(len(coef_r), d=1 / fs_actual)[:len(coef_r) // 2]

        plt.subplot(num_niveles, 2, 2 * i + 1)
        plt.plot(freqs, fft_r, label='Relaxed', alpha=0.7)
        plt.plot(freqs, fft_s, label='Stress', alpha=0.7)
        plt.title(f'{niveles[idx]} - FFT')
        plt.xlabel("Frecuencia (Hz)")
        plt.grid(True)
        plt.legend()

        # PSD
        nperseg_r = min(256, len(coef_r))
        nperseg_s = min(256, len(coef_s))
        freqs_r, psd_r = welch(coef_r, fs=fs_actual, nperseg=nperseg_r)
        freqs_s, psd_s = welch(coef_s, fs=fs_actual, nperseg=nperseg_s)

        plt.subplot(num_niveles, 2, 2 * i + 2)
        plt.semilogy(freqs_r, psd_r, label='Relaxed', alpha=0.7)
        plt.semilogy(freqs_s, psd_s, label='Stress', alpha=0.7)
        plt.title(f'{niveles[idx]} - PSD (Welch)')
        plt.xlabel("Frecuencia (Hz)")
        plt.ylabel("PSD [V²/Hz]")
        plt.grid(True)
        plt.legend()

    plt.tight_layout()
    plt.suptitle(f'FFT y PSD - Subject {str(subject).zfill(2)} - Canal: {canal} (Parte {parte})', fontsize=16, y=1.02)

    filename = f'Sujeto_{str(subject).zfill(2)}_Canal_{canal.replace(" ", "_")}_FFT_PSD_Parte_{parte}.png'
    os.makedirs(output_dir, exist_ok=True)
    path = os.path.join(output_dir, filename)
    plt.savefig(path, bbox_inches='tight', dpi=300)
    print(f">> Imagen guardada en: {path}")
    plt.close()

def wavelet_fft_psd_plot_split(signal_relaxed, signal_stress, subject, canal, fs=500):
    coef_relaxed = pywt.wavedec(signal_relaxed, 'db4', level=8)
    coef_stress = pywt.wavedec(signal_stress, 'db4', level=8)
    niveles = ['cA8', 'cD8', 'cD7', 'cD6', 'cD5', 'cD4', 'cD3', 'cD2', 'cD1']
    output_dir = r'C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Imagenes_EEG\Sujeto 10'

    # Parte 1: cA8 a cD6 → índices 0 a 3
    plot_fft_psd_segment(coef_relaxed, coef_stress, niveles, canal, subject, fs,
                         idx_start=0, idx_end=4, parte=1, output_dir=output_dir)

    # Parte 2: cD5 a cD1 → índices 4 a 8
    plot_fft_psd_segment(coef_relaxed, coef_stress, niveles, canal, subject, fs,
                         idx_start=4, idx_end=9, parte=2, output_dir=output_dir)
wavelet_fft_psd_plot_split(relaxed_df[canal].values, stress_df[canal].values, subject, canal)
```

## 8. Extracción de características

```python
# Rutas
RELAXED_PATH = r"C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Relaxed"
STRESS_PATH = r"C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Stress"
OUTPUT_PATH = r"C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Feature_Extraction_v2.csv"

# Parámetros
fs = 500
wavelet = 'db4'
niveles = 8
nombres_coef = ['cA8'] + [f'cD{i}' for i in range(8, 0, -1)]
canales = ['EEG Fp1', 'EEG Fp2', 'EEG F3', 'EEG F4', 'EEG F7', 'EEG F8',
           'EEG T3', 'EEG T4', 'EEG C3', 'EEG C4', 'EEG T5', 'EEG T6',
           'EEG P3', 'EEG P4', 'EEG O1', 'EEG O2', 'EEG Fz', 'EEG Cz',
           'EEG Pz', 'EEG A2-A1']
excluir = {'Subject04_1.csv', 'Subject04_2.csv', 'Subject31_1.csv', 'Subject31_2.csv'}

# --- Cálculo de métricas desde histograma del PSD ---
def calcular_métricas_histograma(psd, bins=50):
    # Estimamos una distribución de probabilidad discreta del PSD
    hist, _ = np.histogram(psd, bins=bins, density=True)
    hist = hist + 1e-10  # Evitar problemas numéricos con log(0)
    prob = hist / np.sum(hist)  # Aseguramos suma = 1 para entropía
    mediana = np.median(hist)
    rms = np.sqrt(np.mean(hist**2))
    curt = kurtosis(hist)
    norma = np.linalg.norm(hist)
    ent = entropy(prob)  # Entropía de Shannon
    return mediana, rms, curt, norma, ent

# --- Procesamiento por condición ---
def procesar_condicion(ruta, condicion):
    resultados = []
    for archivo in os.listdir(ruta):
        if not archivo.endswith('.csv') or archivo in excluir:
            continue

        sujeto = archivo.replace('_1.csv', '').replace('_2.csv', '')
        df = pd.read_csv(os.path.join(ruta, archivo), sep=';')

        for canal in canales:
            if canal not in df.columns:
                continue

            señal = df[canal].dropna().values
            coeficientes = pywt.wavedec(señal, wavelet=wavelet, level=niveles)

            for nombre, coef in zip(nombres_coef, coeficientes):
                nivel = 8 if nombre == 'cA8' else int(nombre[2])
                fs_actual = fs / (2 ** nivel)

                # PSD de los coeficientes wavelet
                _, psd = welch(coef, fs=fs_actual, nperseg=256)

                # Cálculo de métricas desde el histograma del PSD
                mediana, rms, curt, norma, entropia = calcular_métricas_histograma(psd)

                resultados.append([
                    sujeto, canal, nombre, mediana, rms, curt, norma, entropia, condicion
                ])
    return resultados

# --- Ejecución para ambas condiciones ---
res_relaxed = procesar_condicion(RELAXED_PATH, 0)
res_stress = procesar_condicion(STRESS_PATH, 1)

# --- Guardado de resultados ---
columnas = ['Sujeto', 'Canal', 'Coeficiente', 'Mediana', 'RMS', 'Curtosis', 'Norma', 'Entropia', 'Condicion']
df_final = pd.DataFrame(res_relaxed + res_stress, columns=columnas)
df_final.to_csv(OUTPUT_PATH, sep=';', index=False, float_format='%.6f')

print("✅ Archivo generado correctamente en:")
print(OUTPUT_PATH)
```

```python
warnings.filterwarnings('ignore')

# === Cargar datos ===
ruta_csv = r"C:\Users\Lenovo\OneDrive\Escritorio\IEEE-SP\Arithmetic Project\Resultados\Pivot_Feature_v2.csv"
df = pd.read_csv(ruta_csv, sep=";")
df.columns = df.columns.str.strip()
X = df.drop(columns=["Sujeto", "Condicion"])
y = df["Condicion"]

# Preprocesamiento
X = X.apply(pd.to_numeric, errors='coerce').dropna()
y = y.loc[X.index]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# === Función de evaluación (para PSO) ===
def objective(params, X, y, kernel, degree_fixed=3):
    C = params[0]
    gamma = params[1] if kernel in ['rbf', 'sigmoid'] else 'scale'
    degree = int(params[2]) if kernel == 'poly' else degree_fixed

    svc = SVC(kernel=kernel, C=C, gamma=gamma, degree=degree)
    pipeline = make_pipeline(StandardScaler(), svc)
    score = cross_val_score(pipeline, X, y, cv=5, scoring='accuracy').mean()
    return -score  # Negamos porque PSO minimiza

# === Configuración PSO por kernel ===
kernel_spaces = {
    'linear': ([0.01], [100]),                         # Solo C
    'rbf': ([0.01, 1e-4], [100, 1]),                   # C, gamma
    'sigmoid': ([0.01, 1e-4], [100, 1]),               # C, gamma
    'poly': ([0.01, 1e-4, 2], [100, 1, 5])             # C, gamma, degree
}

resultados = []

for kernel, (lb, ub) in kernel_spaces.items():
    dim = len(lb)
    print(f"\n🚀 Optimizando kernel: {kernel}")

    best_params, _ = pso(
        func=objective,
        lb=lb,
        ub=ub,
        args=(X_train, y_train, kernel),
        swarmsize=20,
        maxiter=10,
        debug=False
    )

    # Ajustar el modelo final con mejores hiperparámetros
    C = best_params[0]
    gamma = best_params[1] if kernel in ['rbf', 'sigmoid', 'poly'] else 'scale'
    degree = int(best_params[2]) if kernel == 'poly' else 3

    clf = make_pipeline(StandardScaler(), SVC(kernel=kernel, C=C, gamma=gamma, degree=degree))
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, pos_label=1)
    precision = precision_score(y_test, y_pred, pos_label=1)
    recall = recall_score(y_test, y_pred, pos_label=1)

    print(f"✅ Mejor C: {C:.4f} | gamma: {gamma} | degree: {degree if kernel == 'poly' else 'N/A'}")
    print(f"📊 Accuracy: {acc:.4f} | F1-score: {f1:.4f} | Precision: {precision:.4f} | Recall: {recall:.4f}")

    resultados.append({
        'Kernel': kernel,
        'C': C,
        'Gamma': gamma,
        'Degree': degree if kernel == 'poly' else None,
        'Accuracy': acc,
        'F1 Score (Stress)': f1,
        'Precision (Stress)': precision,
        'Recall (Stress)': recall
})


# === Mostrar resultados ===
df_resultados = pd.DataFrame(resultados)
print("\n📈 Comparación final por kernel:\n", df_resultados)
```
